```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
article_title = "MQTT"
transversal_level = "1..2"

filename = "results4/" + article_title + "-" + transversal_level + ".csv"

data = pd.read_csv(filename, header=1, index_col=0)

data.head()
```

```python
#for i in range(10):
    #kmeans = KMeans(n_clusters=5, random_state=0).fit(data)
    #print kmeans.labels_
    #print "\n"

kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
#df2 = pd.DataFrame(kmeans.labels_[:,None], columns=["labels"])
```

```python
data['labels'] = pd.Series(kmeans.labels_, index=data.index)
data.head()
#print kmeans.predict(data)
```
